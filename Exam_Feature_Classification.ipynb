```python
#import all required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Read given csv file
data=pd.read_csv('/content/IRIS.csv')
```

```python
#find dimensions of the data, number of columns and rows
data.shape
```

```python
#know first five entries
data.head()
```

```python
#convert a Pandas DataFrame (or Series) into a NumPy array, x=input, y=output
df=data.values
X=df[:,0:4]
Y=df[:,4]
```

```python
#print input
X
```

```python
#print output
Y
```

```python
#visualize all pairs from the dataset and color them
sns.pairplot(data,hue='species')
```

```python
#divide the dataset into training and testing dataset, here the test datasize is 20%
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
```

```python
#find dimensions of the training input dataset, number of columns and rows
X_train.shape
```

```python
#find dimensions of the testing output dataset, number of columns and rows
Y_test.shape
```

```python
#create and train an SVM (Support Vector Machine) classification model
from sklearn.svm import SVC
svc=SVC()
svc.fit(X_train,Y_train)
```

```python
#predict the flower species using a trained SVM model
x_new=[[6.7,3.3,5.7,2.5]]
answer=svc.predict(x_new)
print("For given data, flower is of type= ", answer)
```

```python
#Use the trained SVM model to predict outputs/classes for test data.
y_predict=svc.predict(X_test)
```

```python
#calculate the accuracy of the Machine Learning model
from sklearn.metrics import accuracy_score
acc=accuracy_score(Y_test,y_predict)
print('Accuracy of the model is=',acc)
```

```python
#calculate confusion matrix
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(Y_test,y_predict)
print("Confusion matrix:\n",cm)
```

```python
#generate classification report
from sklearn.metrics import classification_report
print("Classification report\n")
print(classification_report(Y_test,y_predict))
```
