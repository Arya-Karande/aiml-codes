```python
#convert text into numerical form
cv=CountVectorizer()
x_train_count=cv.fit_transform(x_train.values)
```

```python
#train a Naive Bayes Machine Learning model to detect spam/ham mail
model= MultinomialNB()
model.fit(x_train_count, y_train)
```

```python
#predict the cases, whether they are spam or ham
#case - 1
case_1=['Hello students, bring prints for the practical, attached to this mail']
case_1_count=cv.transform(case_1)
model.predict(case_1_count)
```
