```python
#import all required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Read given csv file
data = pd.read_csv('#paste path here')
```

```python
#extracts a signal column from the dataset and calculates the number of data points in the signal
signal=data['mm'].values
n = len(signal)
n
```

```python
#sampling frequency
fs = 1000
dt = 1/fs
dt
```

```python
#Creating time axis
time = np.arange(0, n*dt, dt)
time
```

```python
#plot the vibration signal in the time domain using Matplotlib
plt.figure()
plt.plot(time, signal)
plt.title("Engine Vibration - Time Domain")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid()
plt.show()
```

```python
#perform Fast Fourier Transform analysis to convert the vibration signal from the time domain into the frequency domain
fft_values = np.fft.fft(signal)
freqs = np.fft.fftfreq(n, d=dt)
```

```python
#extract the positive frequency components from the FFT result and calculates the corresponding amplitude spectrum
freqs = freqs[:n//2]
magnitude = np.abs(fft_values[:n//2]) * (2/n)
```

```python
#plot the frequency spectrum of the vibration signal after performing Fast Fourier Transform analysis
plt.figure()
plt.plot(freqs, magnitude)
plt.title("Engine Vibration - Frequency Spectrum")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude")
plt.grid()
plt.show()
```

```python
#find and display the top 5 dominant frequencies present in the vibration signal after performing Fast Fourier Transform analysis
top_indices = magnitude.argsort()[-5:]

print("Top 5 Frequencies:")

for i in top_indices:
    f = freqs[i]
    m = magnitude[i]
    print(f"Frequency: {f:.2f} Hz, Amplitude: {m:.4f}")
```
