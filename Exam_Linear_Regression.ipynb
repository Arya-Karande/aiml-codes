```python
#import all required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
#Read given csv file
data=pd.read_csv('#paste path here')
```

```python
#know first five entries
data.head()
```

```python
#function used to generate summary statistics for all columns in a dataset
data.describe(include='all')
```

```python
#used in Machine Learning to separate the input feature and target/output from a dataset
x=data['Hours','Internal Marks','Attendance']
y=data['Scores']
```

```python
#scatter plot for hours and scores on x & y axis respectively
plt.scatter(x,y)
plt.title("Hours vs Scores")
plt.xlabel("Hours")
plt.ylabel("Scores")
plt.show()
```

```python
#convert the data into a 2-dimensional form, which is required by many models in scikit-learn
X=x.values.reshape(-1,1)
Y=y.values.reshape(-1,1)
```

```python
#divide the dataset into training and testing dataset, here the test datasize is 25% & training is 75% (as testing size is not specified separately)
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y)
```

```python
#creates and trains a Linear Regression model using scikit-learn
reg=LinearRegression()
reg.fit(x_train,y_train)
```

```python
#code is used to visualize the Linear Regression model by plotting
line=reg.coef_*X+reg.intercept_
plt.scatter(X,Y)
plt.plot(X,line,c='red')
plt.title("Linear Regression Output")
plt.show()
```

```python
#predict output values using the trained Linear Regression model
y_predict=reg.predict(x_test)
```

```python
#below are the performance metrics
```

```python
#calculate the Mean Square Error
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y_test,y_predict)
rmse=mse**(1/2)
print('Mean Squared Error is=',mse)
print('Root Mean Squared Error is= ',rmse)
```

```python
#calculate the Mean Absolute Error
from sklearn.metrics import mean_absolute_error
mae=mean_absolute_error(y_test,y_predict)
print('Mean Absolute Error is=',mae)
```
