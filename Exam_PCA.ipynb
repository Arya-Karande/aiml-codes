```python
#import all required libraries
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
#Read given csv file
data=pd.read_csv("#paste path here")
```

```python
#find dimensions of the data, number of columns and rows
data.shape
```

```python
#check the number of missing values in each column of the dataset using Pandas
print(data.isnull().sum())
```

```python
#drop missing values
data = data.dropna()
```

```python
#select only the numerical columns from a dataset using Pandas
X=data.select_dtypes(include=["float64","int64"])
```

```python
#standardize the numerical data using scikit-learn before applying machine learning algorithms
scaler= StandardScaler()
X_scaled=scaler.fit_transform(X)
```

```python
#apply Principal Component Analysis (PCA) to reduce the dataset into 2 principal components
pca=PCA(n_components=2)
X_pca=pca.fit_transform(X_scaled)
```

```python
#display the shape (dimensions) of the PCA-transformed dataset
X_pca.shape
```

```python
#shows how much information each principal component retains
print(pca.explained_variance_ratio_)
```

```python
#visualize the Principal Component Analysis output using a scatter plot
plt.scatter(X_pca[:,0],X_pca[:,1])
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.title("PCA for Air Pollution Data")
plt.grid()
plt.show()
```

```python
#decides optimal PCA components
import numpy as np

cumulative = np.cumsum(pca.explained_variance_ratio_)

plt.plot(cumulative)
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Explained Variance")
plt.grid()
plt.show()
```
